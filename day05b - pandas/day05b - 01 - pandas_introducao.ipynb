### Criando método

```python
def plot_e_salva(df, colx, coly, xlabel, ylabel, title, filefig,
                 root_figure = "../figure/", figsize=(7,5), dpi=60):
    
    fig = plt.figure(figsize=figsize)
    
    plt.plot(df[colx], df[coly])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    if not os.path.exists(root_figure):
        os.mkdir(root_figure)

    filename = os.path.join(root_figure, filefig)
    
    try:
        plt.savefig(filename, dpi=dpi)
        print(f"Figura salva: '{filename}'")
        ret = True
    except:
        print(f"Erro ao salvar figura salva: '{filename}'")
        ret = False

    ret
```

```python
plot_e_salva(dfmycars2, colx='cil', coly='autonomia', 
             xlabel="cilindros", ylabel="autonomia (km)",
             title = "meus carros", filefig='meu_carro.png',
             root_figure = "../figure/", figsize=(7,5), dpi=60)
```

### Melhorando o plot, controlando fontes (tipo de lebra, tamanho e cor)

```python
font_title = {'family': 'serif',
              'color':  'darkred',  # vermelho escuro
              'size': 20,
             }

font_xlab = {'family': 'serif',
             'color':  'navy',   # navy .... azul escuro / naval
             'size': 16,
            }

font_ylab = {'family': 'serif',
             'color':  'blue',
             'size': 16,
            }
try:
    os.mkdir(root_figure)
except:
    pass

figname = "carro_rendimento_por_cilindro_better_font.png"
filename = os.path.join(root_figure, figname)

fig = plt.figure(figsize=(7,5))

plt.plot(dfmycars2.cil, dfmycars2.autonomia)
plt.xlabel("cilindros", fontdict=font_xlab)
plt.ylabel("autonomia (km)", fontdict=font_ylab)
plt.title("meus carros - adoro fuskinha", fontdict=font_title);
plt.savefig(filename, dpi=60);
```
