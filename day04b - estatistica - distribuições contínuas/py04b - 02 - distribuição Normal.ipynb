```python
from scipy import stats
import numpy as np
import pandas as pd

from statistics import mode

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Distribui√ß√£o Normal

**Probabilistic Density Function**

<font size="6">Normal PDF = f(x) = $\frac{1}{\sqrt{2\pi*SSD^2}}*\exp{(\frac{-(x-<x>)^2}{2*SSD^2})}$</font>

https://en.wikipedia.org/wiki/Normal_distribution

#### Analisando distribui√ß√£o normal com seaborn
  - Distribui√ß√£o cont√≠nua
  - Logo, utiliza vari√°veis continuas, exemplo
    - altura e peso de pessoas
    - temperatura em pessoas saud√°veis
    - densidade de um composto
    - m√©dia de acidentes numa via !!!

#### Critique
  - Distribui√ß√£o cont√≠nua e normal
    - altura e peso de pessoas
    - m√©dia de acidentes numa via !!!

As distribui√ß√µes acima podem divergir da normal?

### Simulando uma distribui√ß√£o normal 

#### Amostrando com np.random.normal()

 - media - mean - loc
 - desvio padr√£o - scale

```python
# simula√ß√£o de altura
MU = 170  # cm
SSD = 12
N = 10000

title = f"Uma distribui√ß√£o normal media={MU}({SSD}) cm e com {N} valores"

vals = np.random.normal(MU, SSD, N)

#- mu = media calculada
mu = np.mean(vals)
ssd = np.std(vals)
med = np.median(vals)
mod = mode(vals)

# distplot do seaborn = histograma
ax = sns.displot(vals, height=4, aspect=2)
plt.title(title);
```

### estudos acima de media + 2*SD e abaixo media - 2*SD

  - acima = 170 + 2*15 = 200 cm
    - acima de 200 (media + 2*SD) --> area 2.5% == p-valor = 2.5%
  - abaixo = 170 - 2*15 = 140 cm
    - acima de 200 (media - 2*SD) --> area 2.5% == p-valor = 2.5%


```python
sns.displot(vals, height=4, aspect=2)

#-- vertical line
plt.axvline(mu, color='red')
#- MU = media definida e colocada na maquina np.random.normal(MU ....)
plt.axvline(med, color='orange')

#- dispers√£o: Desvio Padr√£o
plt.axvline(mu+ssd, color='brown')
plt.axvline(mu-ssd, color='brown')
plt.title(title);
```

### Variando m√©dias e desvios padr√µes

```python
mu_params = [-1, 0, 1]
sd_params = [0.5, 1, 1.5]

x = np.linspace(-6, 6, 100)
f, ax = plt.subplots(len(mu_params), len(sd_params), sharex=True, sharey=True, figsize=(12,8))

for i in range(3):
    for j in range(3):
        mu = mu_params[i]
        sd = sd_params[j]
        y = stats.norm(mu, sd).pdf(x)
        ax[i,j].plot(x, y)
        ax[i,j].plot(0, 0, label=f"mu = {mu:3.2f}\nSSD = {sd:3.2f}", alpha=0)
        ax[i,j].legend(fontsize=12);

ax[2,1].set_xlabel('$x$', fontsize=16);
ax[1,0].set_ylabel('$pdf(x)$', fontsize=16);

plt.tight_layout()
```

### Dist√¢ncia entre distribui√ß√µes - prepara√ß√£o para o teste-t

```python
MU_list = [1, 3]
SD_list = [0.3, 0.8]
N = 1000

samp1 = np.random.normal(MU_list[0], SD_list[0], N)
samp2 = np.random.normal(MU_list[1], SD_list[1], N)

#-- dataframe - distribui√ß√£o 1 coluna dist1 e ...
df = pd.DataFrame([samp1, samp2]).T
df.columns=['dist1', 'dist2']
df.head(3)
```

### Duas distribui√ß√µes sobrepostas

```python
colors = ['red', 'blue']

fig, ax = plt.subplots(figsize=(12, 6))
# fig = plt.figure(figsize=(12, 6))


for i in range(2):
    # histogram 
    col = 'dist%d'%(i+1)
    sns.histplot(df[col], stat='density', color=colors[i], fill=True, alpha=0.5, label=col, ax=ax)
    # rug
    sns.rugplot(df[f'dist{i+1}'], color=colors[i], alpha=0.1, height=0.08, ax=ax)

mu1 = df.dist1.mean()
sd1 = df.dist1.std()

mu2 = df.dist2.mean()
sd2 = df.dist2.std()

    
title  =   "Distribui√ß√£o 1 (vermelha) tem m√©dia = %.2f e SSD = %.2f"%(mu1, sd1)
title += "\nDistribui√ß√£o 2 (azul) tem m√©dia = %.2f e SSD = %.2f"%(mu2, sd2)
plt.legend()
plt.title(title);
```

<font size="5">$\theta = \frac{media_1 - media_2}{SD}$</font>

### Tambanho de Efeito de Cohen

```python
# desvio padr√£o conjunto

import math

N1 = N
N2 = N

SDpool = math.sqrt( ((N1-1)*SSD1**2 + (N2-1)*SSD2**2) / (N1+N2-2) )

f"SSD1 = {SSD1:.2f}, SSD2 = {SSD2:.2f} e SDpool = {SDpool:.2f}"
```

```python
diff = (MU2 - MU1) / SDpool

# effect size - tamanho de efeito
f"MU2 = {MU2:.2f}, MU1 = {MU1:.2f} e SDpool = {SDpool:.2f} -> ES = {diff:.2f}"
```

### Spoiler - teste de hip√≥tese
#### Em breve teste de hip√≥tese: s√£o as duas distribui√ß√µes diferentes?

```python
colors = ['blue', 'red']

fig, ax = plt.subplots(figsize=(12, 6))

for i in range(2):
    col = 'dist%d'%(i+1)
    ax = sns.histplot(x=df[col], stat='density',  color=colors[i], ax=ax)

    val_inf = stats.norm.ppf(0.001, MU_list[i], SD_list[i])
    val_sup = stats.norm.ppf(0.999, MU_list[i], SD_list[i])
    
    seqx = np.linspace(val_inf, val_sup, 100)
    pdf_norms = stats.norm.pdf(seqx, MU_list[i], SD_list[i])

    sns.lineplot(x=seqx, y=pdf_norms, color=colors[i])

mu1 = df.dist1.mean()
sd1 = df.dist1.std()

mu2 = df.dist2.mean()
sd2 = df.dist2.std()

plt.axvline(x=MU_list[0], ymin=0, ymax=1.4, color='black')
plt.axvline(x=MU_list[0]+SD_list[0], ymin=0, ymax=1.0, color='black', linestyle='--')
plt.axvline(x=MU_list[0]-SD_list[0], ymin=0, ymax=1.0, color='black', linestyle='--')

plt.axvline(x=MU_list[1], ymin=0, ymax=.6, color='black')
plt.axvline(x=MU_list[1]+SD_list[1], ymin=0, ymax=.4, color='black', linestyle='--')
plt.axvline(x=MU_list[1]-SD_list[1], ymin=0, ymax=.4, color='black', linestyle='--')

head_width = 0.05

# nsamp = N*sqrt(SD_list[0]**2/n**2 + SD_list[1]**2/n**2)
n1=N
n2=N

SDpool = math.sqrt( ((n1-1)*sd1**2 + (n2-1)*sd2**2) / (n1+n2-2) )
EffSize = (mu2 - mu1) / SDpool

diff = MU_list[1]-MU_list[0]
xdiff = MU_list[1] - (diff/2)
yarrow = 0.7

sta, pval = stats.ttest_ind(df.dist1, df.dist2)
#text_stat = "t-test statistics = %.2f, p-val = %.2e"%(sta, pval)
text_stat = "estat√≠stica t-test = %.2f, p-val = %.2e"%(sta, pval)
    
plt.arrow(MU_list[0], yarrow, (diff-2*head_width), 0, head_width=head_width, head_length=0.1, fc='k', ec='k')

text = 'ES = %.2f, diff = %.1f, SSD = %.2f'%(EffSize, diff, SDpool)
plt.text(x=(xdiff-.65), y=(yarrow+.1), s=text)
plt.xlabel("values")
plt.ylabel("percentage (%)")
     
title  =   "Distribui√ß√£o 1 (azul) tem media (SSD) = %.1f (%.2f) e n = %d amostras"%(mu1, sd1, N)
title += "\nDistribui√ß√£o 2 (vermelha) tem media (SSD) = %.1f (%.2f) e n = %d amostras"%(mu2, sd2, N)
title += '\n' + text_stat + '\nTamanho do efeito = %.2f'%(EffSize)
plt.title(title);
```
