### Distribui√ß√£o Normal

**Probabilistic Density Function**

<font size="6">Normal PDF = f(x) = $\frac{1}{\sqrt{2\pi*SSD^2}}*\exp{(\frac{-(x-<x>)^2}{2*SSD^2})}$</font>

https://en.wikipedia.org/wiki/Normal_distribution

### Simulando uma distribui√ß√£o normal

#### Amostrando com np.random.normal()

 - media - mean - loc
 - desvio padr√£o - scale

```python
# simula√ß√£o de altura
MU = 170  # cm
SSD = 12
N = 10000

title = f"Uma distribui√ß√£o normal media={MU}({SSD}) cm e com {N} valores"

vals = np.random.normal(MU, SSD, N)

#- mu = media calculada
mu = np.mean(vals)
ssd = np.std(vals)
med = np.median(vals)
mod = mode(vals)

sns.displot(vals, height=4, aspect=2)

#-- vertical line
plt.axvline(mu, color='red')
#- MU = media definida e colocada na maquina np.random.normal(MU ....)
plt.axvline(med, color='orange')

#- dispers√£o: Desvio Padr√£o
plt.axvline(mu+ssd, color='brown')
plt.axvline(mu-ssd, color='brown')
plt.title(title);
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(3, 1, figsize=(10,8), sharex=True)
sns.despine(left=True)

#------------- eixo 0 - pdf -------------------
ax=axes[0]
ax = sns.histplot(vals, kde=False, color='blue', ax=ax)
ax.set_ylabel("Contagem")
ax.set_title("Gr√°fico de Frequ√™ncia - dados uniformemente amostrados")

ax.set_title(title)

ax.axvline(mu, color='red')
ax.axvline(med, color='orange')
ax.axvline(mu+ssd, color='brown')
ax.axvline(mu-ssd, color='brown')

# Add annotation
outlier_inf = 130
ax.annotate(
    'Outliers Inf',
    xy=(outlier_inf, 0),  # Point to annotate (arrow tip)
    xytext=(outlier_inf, 200),  # Position of the text
    arrowprops=dict(facecolor='black', shrink=0.05), # Arrow properties
    horizontalalignment='left', # Text alignment
    verticalalignment='center' # Text alignment
)

# Add annotation
outlier_sup = 210
ax.annotate(
    'Outliers Sup',
    xy=(outlier_sup, 0),  # Point to annotate (arrow tip)
    xytext=(outlier_sup, 200),  # Position of the text
    arrowprops=dict(facecolor='black', shrink=0.05), # Arrow properties
    horizontalalignment='left', # Text alignment
    verticalalignment='center' # Text alignment
)

#------------- eixo 1 - boxplot -------------------
ax=axes[1]
ax = sns.boxplot(x=vals, color='lightblue', ax=ax)
ax.set_title("Boxplot - q1 <> mu-SSD e q2 <> mu+SSD")

ax.axvline(mu, color='red')
ax.axvline(med, color='orange')
ax.axvline(mu+ssd, color='brown')
ax.axvline(mu-ssd, color='brown')

#------------- eixo2 - cumulativa -------------------
ax=axes[2]
# ecdf - cumulative
sns.ecdfplot(x=vals, color="steelblue", ax=ax)

ax.axvline(mu, color='red')
ax.axvline(med, color='orange')
ax.axvline(mu+ssd, color='brown')
ax.axvline(mu-ssd, color='brown')

ax.set_title('cumulativa');
```

### Distribui√ß√£o t de William Sealy Gosset

Distribui√ß√£o t para n-1 graus de liberdade (dof)

<font size="6">t-stat = $\frac{<X>-\mu}{\frac{SSD}{\sqrt(n)}}$ ~ $t_{n-1}$</font>

https://en.wikipedia.org/wiki/Student%27s_t-distribution

### Distribui√ß√£o t-student

```python
from scipy.stats import t
```

üìê Skewness (assimetria) - terceiro momento
üìå Defini√ß√£o

Skewness mede assimetria da distribui√ß√£o:

  - 0 ‚Üí sim√©trica
  - \> 0 ‚Üí cauda longa √† direita
  - < 0 ‚Üí cauda longa √† esquerda

üìä Kurtosis (peso das caudas) - quarto momento  
‚ö†Ô∏è Aten√ß√£o: SciPy usa excess kurtosis

  - Normal ‚Üí kurtosis = 0
  - Leptoc√∫rtica (caudas pesadas) ‚Üí \> 0
  - Platic√∫rtica (caudas leves) ‚Üí < 0

```python
degfree=3

float(t.ppf(0.01, degfree))
```

```python
x = np.linspace(t.ppf(0.01, degfree), t.ppf(0.99, degfree), 100)
plt.plot(x, t.pdf(x, degfree), 'r-', lw=3, alpha=0.6, label='t pdf');
```

```python
colors = ['blue', 'tomato', 'green', 'violet']
# dof - degree of freedom - graus de liberdade
# N < 100 .... N-1 = dof
# dist normal --> mas, 6 camundongos --> t-student ou nao-param√©trica
degfrees = [3, 8, 30, 1000]

plt.figure(figsize=(12,6))

for i in range(4):
    color = colors[i]
    degfree    = degfrees[i]

    x = np.linspace(t.ppf(0.001, degfree), t.ppf(0.999, degfree), 100)
    plt.plot(x, t.pdf(x, degfree), '-', lw=3, alpha=0.6, label='t degfree=%d'%degfree, color=color);

seqx = np.linspace(-6, 6, 100)
plt.plot(seqx, stats.norm.pdf(seqx, 0, 1), color='black', label='normal dist.')
plt.title("t-distribution & normal distribution (in black)")

plt.xlim(-6,6)
plt.legend();
```

```python
#-- degree of freedom - graus de liberdade
#-- se fa√ßo um experimnto N = 3 ... degree of freedom = N-1

print("dof = degree of freedom = graus de liberdade")
print("skewness = assimetria")
print("kurtosis = curtose - achatamento da curva\n\n")

print("De acordo com que o dof cresce a distr.student tende √† normal\n")

for degfree in degfrees:
    mean, var, skew, kurt = t.stats(degfree, moments='mvsk')
    msg = f"graus de liberdade (dof) = {degfree:4}, m√©dia={mean:.2f} var={var:.2f} skewness={skew:.3f} kurtosis={kurt:.3f}"
    print(msg)
```
