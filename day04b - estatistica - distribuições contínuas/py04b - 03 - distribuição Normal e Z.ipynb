````python
from scipy import stats
import numpy as np
import pandas as pd

from statistics import mode

import seaborn as sns
import matplotlib.pyplot as plt
# %matplotlib inline
````

### Distribui√ß√£o Normal

**Probabilistic Density Function**

<font size="6">Normal PDF = f(x) = $\frac{1}{\sqrt{2\pi*SSD^2}}*\exp{(\frac{-(x-<x>)^2}{2*SSD^2})}$</font>

https://en.wikipedia.org/wiki/Normal_distribution

### Simulando uma distribui√ß√£o normal

#### Amostrando com np.random.normal()

 - media - mean - loc
 - desvio padr√£o - scale

````python
# simula√ß√£o de altura
MU = 170  # cm
SSD = 12
N = 10000

title = f"Uma distribui√ß√£o normal media={MU}({SSD}) cm e com {N} valores"

samples = np.random.normal(MU, SSD, N)

#- mu = media calculada
mu = np.mean(samples)
ssd = np.std(samples)
med = np.median(samples)
mod = mode(samples)

sns.displot(samples, height=4, aspect=2)

#-- vertical line
plt.axvline(mu, color='red')
#- MU = media definida e colocada na maquina np.random.normal(MU ....)
plt.axvline(med, color='orange')

#- dispers√£o: Desvio Padr√£o
plt.axvline(mu+ssd, color='brown')
plt.axvline(mu-ssd, color='brown')

plt.axvline(mu+2*ssd, color='black')
plt.axvline(mu-2*ssd, color='black')


plt.title(title);
````

### O que √© uma distribui√ß√£o Z

#### uma distribui√ß√£o Z √© uma distribui√ß√£o centrada em 0 e com desvio padr√£o = 1

ou seja √© uma **distribui√ß√£o normalizada**

  - centrada em zero
  - desvio padr√£o = 1

<br>

<font size="5">Z = $\frac{x - <x>}{SSD}$</font>

<br>

````python
z_vals = (samples-mu)/SSD

muZ = np.mean(z_vals)  # = 0
SSDZ = np.std(z_vals)  # = 1

fig, ax = plt.subplots(figsize=(12, 6))

sns.set_theme(style="darkgrid")

ax = sns.histplot(x=z_vals, stat='density',  color="steelblue", ax=ax)

ax.vlines(0, 0, 0.42, color = 'black');
ax.hlines(.24, 0, 0+SSDZ, color = 'red');

#--- fitando a distribui√ß√£o normal com mu e SSD
#---                          Percent Point Function (PPF)
# seqx = np.linspace(stats.norm.ppf(0.01, muZ, SSDZ), stats.norm.ppf(0.99, muZ, SSDZ), 100)

std_inf = -4
std_sup = +4
seqx = np.linspace(std_inf, std_sup, 100)

z_pdf = stats.norm.pdf(seqx, loc=0, scale=1)
sns.lineplot(x=seqx, y=z_pdf, color='navy', ax=ax)

title = f'Distribui√ß√£o Z sempre tem m√©dia=0 (sample mean {muZ:.2f} e desvio padr√£o = 1 (sample SD={SSDZ:.2f})'
plt.title(title);
````

````python
perc25_inf = stats.norm.ppf(0.025, MU, SSD)
perc25_sup = stats.norm.ppf(0.975, MU, SSD)

print("Normal-rand√¥mico")
f"N={N} mu {mu:.3f}, ssd {ssd:.3f}, p25_inf {perc25_inf:.3f}, p25_sup {perc25_sup:.3f}"
````

````python
# go --> plotly

alpha = 0.025

# x range
seqx = np.linspace(MU - 4*SSD, MU + 4*SSD, 2000)
pdf = norm.pdf(seqx, MU, SSD)

# valor de corte = 2.5%
cutoff = norm.ppf(alpha, MU, SSD)

fig = go.Figure()

# --- Normal curve
fig.add_trace(
    go.Scatter(x=seqx, y=pdf,
        mode="lines", line=dict(color="navy", width=2),
        name="~Normal"
    )
)

# --- Left tail (first 2.5%)
mask = seqx < cutoff

fig.add_trace(
    go.Scatter(
        x=seqx[mask], y=pdf[mask],
        mode="lines", fill="tozeroy",
        fillcolor="rgba(0, 0, 255, 0.4)", line=dict(color="lightblue"), name="Abaixo"
    )
)

# --- Lihha de cutoff
fig.add_vline(
    x=cutoff,
    line_dash="dash", line_color="blue",
    annotation_text=f"z = {cutoff:.2f}",
    annotation_position="top left"
)

# --- Layout
fig.update_layout(
    width=900,
    height=500,
    title="Distribui√ß√£o Normal com √°rea pintada < 2.5%",
    xaxis_title="x",
    yaxis_title="Densidade",
)

fig.show()
````

### Como se compara duas distribui√ß√µes Z?

a primeira distribui√ß√£o √© a controle
dela obtemos a m√©dia e o desvio padr√£o

a segunda distribui√ß√£o √© o case
obetmos sua distribui√ß√£o z dividindo seus valores amostrais pela media e desvio padr√£o do controle

````python
MU1 = 2; SSD1 = 1.5
MU2 = 4; SSD2 = 1.8
N   = 1000

samp1 = np.random.normal(loc=MU1, scale=SSD1, size=N)
samp2 = np.random.normal(loc=MU2, scale=SSD2, size=N)
print(np.mean(samp2))

z1 = (samp1-MU1)/SSD1
#-- z2 em rela√ß√£o aos par√¢metros de z1 !!!
z2 = (samp2-MU1)/SSD1

muZ1 = np.mean(z1); ssdZ1 = np.std(z1)
muZ2 = np.mean(z2); ssdZ2 = np.std(z2)

muc1 = np.mean(z1); ssd1 = np.std(z1)
muc2 = np.mean(z2); ssd2 = np.std(z2)

fig, ax = plt.subplots(figsize=(12, 6))

# ax  = sns.distplot(z1, kde=False, rug=True, norm_hist=True, color='blue', label='control', rug_kws={"color": 'blue', "alpha": .1,})
# ret = sns.distplot(z2, kde=False, rug=True, norm_hist=True, color='red',  label='case'   , rug_kws={"color": 'red',  "alpha": .1,}, ax=ax)

ax = sns.histplot(x=z1, stat='density',  color="blue", label='control', alpha=.4, ax=ax)
ax = sns.histplot(x=z2, stat='density',  color="orange",  label='case', alpha=.4, ax=ax)

ax.vlines(  0, 0, 0.42,    color = 'navy')
ax.hlines(.24, 0, 0+ssdZ1, color = 'navy');

ax.vlines(muc2, 0, 0.31, color = 'red', linestyle='--')
ax.hlines(.2, muc2, muc2+ssdZ2, color = 'red', linestyle='--')

ax.vlines(2, 0, 0.1, color = 'black', linestyle='-.')

#--- fitando a distribui√ß√£o normal com mu e SSD
#---                          Percent Point Function (PPF)
# seqx = np.linspace(MU1-4*SSD1, MU2+4*SSD2, 100)
seqx = np.linspace(-5, 9, 100)
sns.lineplot(x=seqx, y=stats.norm.pdf(seqx, muZ1, ssdZ1), color='navy')
sns.lineplot(x=seqx, y=stats.norm.pdf(seqx, muZ2, ssdZ2), color='red')

title = 'Dist√¢ncia entre distribui√ß√µes Z'
title += '\nZ1 tem MU1 = %.2f e ssd1 = %.2f'%(muc1, ssd1)
title += '\nZ2 tem MU2 = %.2f e ssd2 = %.2f'%(muc2, ssd2)
plt.title(title)
plt.legend();
````
